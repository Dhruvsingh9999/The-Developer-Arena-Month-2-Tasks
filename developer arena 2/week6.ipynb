```python
#Probability and Statistical Testing
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
np.random.seed(42)
```

```python
# Simulate 1000 values from a normal distribution
data_normal = np.random.normal(loc=100, scale=15, size=1000)

# Plot the distribution
sns.histplot(data_normal, kde=True, color='skyblue')
plt.title("Normal Distribution (μ=100, σ=15)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Simulate Poisson-distributed data (e.g. daily website visits)
data_poisson = np.random.poisson(lam=10, size=1000)

# Plot the distribution
sns.histplot(data_poisson, bins=20, kde=False, color='orange')
plt.title("Poisson Distribution (λ=10)")
plt.xlabel("Events")
plt.ylabel("Count")
plt.show()
```

```python
# Confidence interval for the mean of the normal dataset
mean = np.mean(data_normal)
std_err = stats.sem(data_normal)
ci = stats.t.interval(0.95, len(data_normal)-1, loc=mean, scale=std_err)

print(f"95% Confidence Interval for the Mean: {ci}")
```

```python
# H0: Mean = 100
t_stat, p_value = stats.ttest_1samp(data_normal, 100)

print(f"T-statistic = {t_stat:.4f}")
print(f"P-value = {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
# Simulate two strategies (e.g. sales data)
strategy_A = np.random.normal(200, 20, 100)
strategy_B = np.random.normal(210, 25, 100)

# Perform independent t-test
t_stat_ab, p_value_ab = stats.ttest_ind(strategy_A, strategy_B)

print(f"T-statistic = {t_stat_ab:.4f}")
print(f"P-value = {p_value_ab:.4f}")

if p_value_ab < 0.05:
    print("Significant difference between Strategy A and B.")
else:
    print("No significant difference between the strategies.")
```

```python
# Create observed frequency table
observed = np.array([[30, 10], [20, 40]])
# Assume columns = [Buy, Not Buy]; rows = [Group A, Group B]

# Chi-squared test
chi2, p, dof, expected = stats.chi2_contingency(observed)

print(f"Chi-squared = {chi2:.4f}")
print(f"P-value = {p:.4f}")
print(f"Expected Frequencies:\n{expected}")
```
