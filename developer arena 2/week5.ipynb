```python
# Week 5 â€“ Exploratory Data Analysis (EDA)
# Dataset: Mock Iris Dataset
# Objective: Summary stats, correlation, visualizations, insights

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style
sns.set(style="whitegrid")

# Create mock Iris-like dataset
np.random.seed(42)
mock_data = pd.DataFrame({
    "sepal_length": np.random.normal(5.8, 0.4, 150),
    "sepal_width": np.random.normal(3.0, 0.3, 150),
    "petal_length": np.random.normal(3.7, 1.5, 150),
    "petal_width": np.random.normal(1.2, 0.5, 150),
    "species": np.random.choice(["setosa", "versicolor", "virginica"], 150)
})

# Preview the data
mock_data.head()
```

```python
# Dataset structure
mock_data.info()

# Descriptive statistics
mock_data.describe()
```

```python
mock_data.isnull().sum()
```

```python
correlation = mock_data.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.pairplot(mock_data, hue='species')
plt.show()

```

```python
sns.boxplot(x='species', y='sepal_length', data=mock_data, palette='Set2')
plt.title('Sepal Length by Species')
plt.show()
```
